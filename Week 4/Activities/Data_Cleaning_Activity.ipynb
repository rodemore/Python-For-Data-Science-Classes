```python
import pandas as pd
```

```python
df = pd.read_csv("mk_data.csv")
df.shape
```

```python
df.columns
```

```python
df["Income"] = df["Income"].str.strip().str.replace(",", "").str.replace("$", "").astype(float)
```

```python
df.describe(include="object")
```

```python
def get_outliers(col): 
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lim_sup = Q3 + 1.5*IQR

    outliers = df[df[col]>lim_sup]
    return outliers
get_outliers("Income")
```

```python
def get_outliers_2(serie): 
    Q1 = serie.quantile(0.25)
    Q3 = serie.quantile(0.75)
    IQR = Q3 - Q1

    lim_sup = Q3 + 1.5*IQR

    outliers = serie[serie>lim_sup]
    return outliers

get_outliers_2(df["Income"])
```
