```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://mysiterobert.s3.us-east-1.amazonaws.com/Dataframes/personas_clean.csv")
df.head(2)
```

```python
df.describe(include="all")
```

```python
#df["edad"].hist(bins = [20, 30, 40, 50]) # Matplotlib
```

```python
sns.histplot(df["edad"], bins = 10, kde=True)
```

```python
sns.histplot(
    data = df, 
    x = "altura_cm", 
    bins = 10,
    kde = True
)
```

```python
df.groupby("sexo")["altura_cm"].median()
```

```python
sns.histplot(
    data = df, 
    x = "altura_cm", 
    hue = "sexo",
    bins = 10,
    kde = True
)
plt.show() # visualizar la gráfica (mostrarla)
```

```python
plt.figure(figsize=(7,3))
sns.boxplot(
    data = df, 
    y = "altura_cm", 
    hue= "sexo", 
    showmeans = True
)

plt.title("Distribucion de altura por género", size = 15)
plt.ylabel("Altura (cm)", size = 12)

plt.savefig("distribuciones.jpg")
```
