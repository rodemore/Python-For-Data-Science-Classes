# 6. Calcular co-ocurrencia de productos

## Concepto

La **co-ocurrencia** mide cuántas veces dos productos aparecen juntos en las mismas órdenes.

Por ejemplo:
- Producto A aparece en órdenes: [1, 2, 3, 5]
- Producto B aparece en órdenes: [2, 3, 4, 5]
- Co-ocurrencia A-B: 3 (aparecen juntos en órdenes 2, 3 y 5)

## ¿Cómo calcularlo?

Queremos crear una tabla donde:
- Cada fila es un producto
- Cada columna es un producto
- El valor en [i, j] = cuántas veces el producto i y el producto j aparecen juntos

## Hints

### Método 1: Multiplicar la matriz por sí misma (recomendado - más rápido)

La matriz que creaste tiene 1s donde el producto está presente en una orden.

**Pasos:**
1. Transponer la matriz: usa `.T` (esto convierte filas en columnas)
2. Multiplicar: transpuesta × matriz original
   - Puedes usar `.dot()` o el operador `@`
   - Ejemplo: `matriz_transpuesta @ matriz_original`
3. Esto te dará una matriz cuadrada (productos × productos)

**¿Por qué funciona?** Cuando multiplicas de esta forma, estás contando cuántas órdenes comparten ambos productos (donde ambos tienen valor 1).

### Método 2: Iterar sobre pares de productos (más lento, pero más intuitivo)

Si el método anterior te parece confuso, puedes hacerlo de forma más directa:

1. Para cada par de productos (A, B):
   - Multiplica la columna del producto A × columna del producto B
   - Suma el resultado (cuenta cuántos 1×1 hay, es decir, órdenes donde ambos están presentes)
   - Guarda este valor como co-ocurrencia de A y B

Este método es más lento pero más fácil de entender.

### Convertir resultado a DataFrame

El resultado de la multiplicación será un array. Conviértelo a DataFrame:
- Usa `pd.DataFrame()`
- Asigna índices y columnas con los nombres/SKUs de productos
- Esto facilita consultas posteriores

### Interpretar la diagonal

**Importante:** La diagonal de la matriz (donde fila = columna) tiene un significado especial:
- Representa cuántas veces cada producto aparece **consigo mismo**
- Es decir: en cuántas órdenes está presente ese producto
- Es la frecuencia individual del producto

**Para análisis de pares:**
- Puedes dejar la diagonal como referencia
- O ponerla en 0 para enfocarte solo en productos diferentes: `np.fill_diagonal(matriz, 0)`
