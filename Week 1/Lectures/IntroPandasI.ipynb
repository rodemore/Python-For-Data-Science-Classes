## 10) Calcular `BMI` (Body Mass Index)

**¿Qué es el BMI?**  
El **Índice de Masa Corporal (BMI)** es una medida aproximada que relaciona peso y altura:

$$
BMI = \frac{peso\,(kg)}{altura\,(m)^2}
$$

**Referencia rápida (adultos):**
- **Sobrepeso**: BMI ≥ 25  
- **Obesidad**: BMI ≥ 30  

**Ejercicio 10:** Crea la columna **`BMI`** usando `peso_kg` y `altura_m`.


```python
# Tu solución aquí
df['BMI'] = df["peso_kg"]/ (df["altura_m"] **2)
df.head(2)
```
