## 4. Análisis RFM

### ¿Qué es RFM?

**RFM** es una técnica de segmentación de clientes basada en tres métricas clave:

- **R (Recency - Recencia):** ¿Cuán recientemente compró el cliente?  
  *Clientes que compraron recientemente tienen mayor probabilidad de volver a comprar.*

- **F (Frequency - Frecuencia):** ¿Con qué frecuencia compra el cliente?  
  *Clientes frecuentes son más leales y valiosos.*

- **M (Monetary - Valor monetario):** ¿Cuánto dinero ha gastado el cliente?  
  *Clientes que gastan más son más rentables.*

### ¿Por qué es importante?

El análisis RFM permite:
- Identificar a los mejores clientes (champions)
- Detectar clientes en riesgo de abandonar
- Personalizar estrategias de marketing
- Optimizar la inversión en retención vs adquisición
- Aumentar el ROI de campañas de marketing
