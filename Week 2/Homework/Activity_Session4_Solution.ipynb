### Ejercicio 2.3: Análisis de Pareto (80/20)

**Instrucciones:**
1. Calcula el porcentaje de ventas que representa cada producto
2. Calcula el porcentaje acumulado
3. Identifica cuántos productos representan el 80% de las ventas

**Pregunta:** ¿Qué porcentaje de productos representa el 80% de las ventas?

```python
# Tu código aquí

ventas_productos = merged.groupby(["sku", "nombre"])["ventas_total"].sum()\
    .reset_index()\
    .sort_values("ventas_total", ascending = False)

ventas_totales = ventas_productos["ventas_total"].sum()
ventas_productos["ventas_pct"] = ventas_productos["ventas_total"] / ventas_totales
```

```python
merged["ventas_total"].min()
```

```python
ventas_productos["ventas_accum"] = ventas_productos["ventas_pct"].cumsum()
ventas_productos[ventas_productos["ventas_accum"]<=0.8]
```

```python
# convertir la fecha en formato datetime
merged["fecha_hora_compra"] = pd.to_datetime(merged["fecha_hora_compra"])
# crear una columna con el mes
merged["mes"] = merged["fecha_hora_compra"].dt.month

merged["ventas_total"] = merged["cantidad"]* merged["precio_unitario"] * (1- merged["descuento_pct"]/100)
```
